```python
# Carregar e exibir a imagem
import urllib.request
import cv2
import numpy as np
from matplotlib import pyplot as plt

# URL da imagem (substitua pelo link correto)
url = "https://www.usmagazine.com/wp-content/uploads/2021/04/The-Big-Bang-Theory-Cast-Where-Are-They-Now-Promo.jpg?crop=0px%2C0px%2C2200px%2C1156px&resize=1200%2C630&quality=40&strip=all"

# Baixar a imagem e salvar no disco
urllib.request.urlretrieve(url, "imagem.jpg")

# Carregar a imagem corretamente
img = cv2.imread("imagem.jpg")

# Verificar se a imagem foi carregada corretamente
if img is None:
    print("Erro ao carregar a imagem. Verifique a URL ou o caminho do arquivo.")
else:
    # Converter para RGB
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Exibir a imagem
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.show()
```

```python
#Instalar as bibliotecas necessárias para a detecção de faces
!pip install mtcnn
!pip install lz4
```

```python
#Detecção de rostos na imagem
from mtcnn import MTCNN

# Inicializar o detector MTCNN
detector = MTCNN()

# Detectar faces na imagem
faces = detector.detect_faces(img_rgb)

# Exibir as faces detectadas
for face in faces:
    x, y, width, height = face['box']
    cv2.rectangle(img_rgb, (x, y), (x + width, y + height), (0, 255, 0), 2)

# Exibir a imagem com as faces detectadas
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
